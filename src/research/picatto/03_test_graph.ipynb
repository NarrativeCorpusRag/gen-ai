```python
from google.cloud import storage
def list_folders(target:str):
    """
    Lists 'subfolders' in a bucket path.
    S3 and GCS simulate folders using delimiters (usually '/').
    """
    client = storage.Client.from_service_account_json('../../../gcp_service_account.json')
    bucket = client.get_bucket('gen-ai-tu')

    blobs = bucket.list_blobs(prefix='', delimiter='/')
    for blob in blobs:
        pass
        
    return target in list(blobs.prefixes)
list_folders('news/')
```

```python

def create_folder(target) -> None:
    client = storage.Client.from_service_account_json('../../../gcp_service_account.json')
    bucket = client.get_bucket('gen-ai-tu')
    blob = bucket.blob(target)
    blob.upload_from_string('')


create_folder('index/test/')
```
